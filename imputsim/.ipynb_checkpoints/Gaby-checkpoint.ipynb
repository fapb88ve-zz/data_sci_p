```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel('Gobierno Digital (Responses).xlsx')
del df['Email address']
df.head()
```

```python
df.columns = ['time', 'sex', 'sanan', 'edad', 'internet', 'fac', 'gobdig', 'efi'
             , 'tramciv', 'recsug', 'treal', 'infas', 'cpar', 'maspar', 'forpar']
```

```python
f, m = df.sex.value_counts()[0], df.sex.value_counts()[1]
tot = len(df.sex)
propf, propm = f/tot, m/tot
```

```python
sexo = stats.bernoulli.rvs(propm, size = 313)
len(sexo)
```

```python
df.edad.value_counts()
```

```python
edad = []
for i in df.edad:
    if i == '18-25':
        edad.append(1)
    elif i == '26-33':
        edad.append(2)
    elif i == '34-41':
        edad.append(3)
    elif i == '42-49':
        edad.append(4)
    elif i == '50-57':
        edad.append(5)
    else:
        edad.append(6)
medad = np.mean(edad)
stdedad = np.std(edad)
q1 = pd.Series(abs(np.random.normal(medad, stdedad, 313).round()))
len(q1)
q1.replace([0,8], medad.round(), inplace = True)
q1.value_counts()
```

```python
inter = []
for i in df.internet:
    if i == 'Sí':
        inter.append(1)
    elif i == 'No':
        inter.append(2)
    elif i == 'Deficiente':
        inter.append(3)
mint = np.mean(inter)
stdint = np.std(inter)
q2 = pd.Series(abs(np.random.normal(mint, stdint, 313).round()))
len(q2)
q2.replace([0, 4], mint.round(), inplace = True)
q2
```

```python
q3s, q3n = df.fac.value_counts()[0], df.fac.value_counts()[1]
props, propn = q3s/tot, q3n/tot
q3 = stats.bernoulli.rvs(props, size = 313)
len(q3)
```

```python
digob = []
for i in df.gobdig:
    if i == 'Sí':
        digob.append(1)
    elif i == 'No':
        digob.append(2)
    elif i == 'He escuchado hablar de el superficialmente':
        digob.append(3)
md = np.mean(digob)
stdd = np.std(digob)
q4 = pd.Series(abs(np.random.normal(md, stdd, 313).round()))
len(q4)
q4.replace([0, 4], md.round(), inplace = True)
q4
```
